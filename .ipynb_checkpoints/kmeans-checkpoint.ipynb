```python
def get_data(file_name = None, representativity = True):
    if not file_name:
        file_in = '../Data/%s.csv' % FILE_NAME
    else:
        file_in = file_name
    egos, typo_vector = [], {}
    nb_par_ego = {}
    with open(file_in, 'r') as to_read:
        csv_r = csv.reader(to_read, delimiter = ';')
        
        first_line = csv_r.next()
        keys_typo = [int(x.split('_')[-1]) for x in first_line[1:len(first_line)]]
        
        for line in csv_r:
            ego = line[0]
            if ego in egos:
                ego = ego + '_%s' % (nb_par_ego.get(ego, 0) + 1)
                nb_par_ego[ego] = 1 + nb_par_ego.get(ego, 0)
            egos.append(ego)
            if not representativity:
                typo_vector[ego] = {int(keys_typo[i]) : float(line[i+1]) for i in range(len(keys_typo))}
            else:
                typo_vector[ego] = \
                {
                    int(keys_typo[i]) : float(line[i+1]) if float(line[i+1]) <= 1 \
                                        else 2-1/float(line[i+1])

                    for i in range(len(keys_typo))
                }
                    
                
    return egos, typo_vector, keys_typo

def get_overall_representativity(graphlets_clusters = None):
    s = {x : 0 for x in DATA_MOTIFS[DATA_MOTIFS.keys()[0]]} if not graphlets_clusters else \
        {x[0] : 0 for x in graphlets_clusters}       
       
    for ego in DATA_MOTIFS:    
        if not graphlets_clusters:  
            for graphlet in DATA_MOTIFS[ego]:
                s[graphlet] += int(DATA_MOTIFS[ego][graphlet])
        else:
            for cluster in graphlets_clusters:
                for graphlet in cluster:
                    s[cluster[0]] += int(DATA_MOTIFS[ego][graphlet])
    
    all_graphlets = sum(s.values())
    return {x : s[x]/float(all_graphlets) for x in s}


def get_classe_representativity(list_of_egos, graphlets_clusters = None):
    s = {x : 0 for x in DATA_MOTIFS[DATA_MOTIFS.keys()[0]]} if not graphlets_clusters else \
        {x[0] : 0 for x in graphlets_clusters}       
    for ego in list_of_egos:
        if not graphlets_clusters:
            for graphlet in DATA_MOTIFS[ego]:
                s[graphlet] += int(DATA_MOTIFS[ego][graphlet])
        else:
            for cluster in graphlets_clusters:
                for graphlet in cluster:
                    s[cluster[0]] += int(DATA_MOTIFS[ego][graphlet])
    
    all_graphlets = float(sum(s.values()))
    local_representativities = {x : s[x]/all_graphlets for x in s}
    
    #if not graphlets_clusters:
    return {x : local_representativities[x] / float(OVERALL_REPR[x]) \
            if local_representativities[x] / float(OVERALL_REPR[x]) < 1
            else
                2 - 1 / (local_representativities[x] / float(OVERALL_REPR[x]))
            for x in local_representativities}
        #return {x : local_representativities[x] / float(OVERALL_REPR[x]) \
        #        if local_representativities[x] / float(OVERALL_REPR[x]) < 1
        #        else
        #            (2 - 1 / float(local_representativities[x])) / (2 - 1 / float(OVERALL_REPR[x]))
        #        for x in local_representativities}
    #else:
    #    return {x : local_representativities[x] / float(OVERALL_REPR_CLUSTERED[x]) \
    #            if local_representativities[x] / float(OVERALL_REPR_CLUSTERED[x]) < 1
    #            else
    #                2 - 1 / (local_representativities[x] / float(OVERALL_REPR_CLUSTERED[x]))
    #            for x in local_representativities}
```

```python
def create_dir(k_value):
    directory = '%s/Typo_%s' % (MAIN_DIR, k_value)
    if not 'Typo_%s' % k_value in os.listdir(MAIN_DIR):
        os.mkdir(directory)
    return directory

def kmeans_all(data):
    
    typo_par_ego_par_k = {ego : {} for ego in EGOS}

    for k_value in K_VALUES :
        s_max = 0
        labels_max = False
        for i in range(100):
            kmeans = KMeans(n_clusters = k_value)
            kmeans.fit(data)
            s = np.average(silhouette_samples(data, kmeans.labels_))
            if s > s_max:
                labels_max = kmeans.labels_
                s_max = s
            
        directory = create_dir(k_value)
        egos_par_classe = {}
        
        members_per_cluster = [[labels_max == k] for k in range(k_value)]
        
        for axe_2 in range(1, len(data.T)):
            for k in range(k_value):
                my_members = members_per_cluster[k][0]
                egos_par_classe[k] = []
                with open('%s/egos_classe_%s.csv' % (directory, k), 'w') as to_write:
                    csv_w = csv.writer(to_write, delimiter = ';')
                    csv_w.writerow(['ego_id'] + KEYS_TYPO)
                    for i in range(len(EGOS)):
                        if my_members[i]:
                            egos_par_classe[k].append(EGOS[i])
                            csv_w.writerow([EGOS[i]] + [TYPO_VECTOR[EGOS[i]][j] for j in KEYS_TYPO])
                            typo_par_ego_par_k[EGOS[i]][k_value] = k

        repr_par_classe, repr_par_classe_clustered = {}, {}
        for classe in egos_par_classe:
            repr_par_classe[classe] = get_classe_representativity(egos_par_classe[classe])
            
        with open('%s/kmeans_stats.csv' % directory, 'w') as to_write:
            csv_w = csv.writer(to_write, delimiter = ';')
            csv_w.writerow(['classe', 'nb', 'prop'] + [""] + KEYS_TYPO)
            classe_par_taille = [classe for classe in egos_par_classe]
            classe_par_taille.sort(key = lambda classe : len(egos_par_classe[classe]))
            for classe in classe_par_taille:
                nb = len(egos_par_classe[classe])
                prop = round(nb/float(len(EGOS)), 2)
                
                row = [classe, nb, prop] + [""]
                for key in KEYS_TYPO:
                    row.append(repr_par_classe[classe][key])                    
                csv_w.writerow(row)      

        print 'silhouette k = %s : %s' % (k_value, round(s_max,3))  
        print kmeans.cluster_centers_

    with open('%s/../typo_par_ego.csv' % directory, 'w') as to_write:
        csv_w = csv.writer(to_write, delimiter =';')
        csv_w.writerow(['ego'] + ['k = %s' % k for k in K_VALUES])
        for ego in typo_par_ego_par_k:
            csv_w.writerow([ego]+[typo_par_ego_par_k[ego][k_v] for k_v in K_VALUES])
```

```python
##Kmeans restreint aux réseaux de certaines tailles

def get_data(ego_size, size_max, file_name = None, representativity = True):
    if not file_name:
        file_in = '../Data/%s.csv' % FILE_NAME
    else:
        file_in = file_name
    egos, typo_vector = [], {}
    nb_par_ego = {}
    with open(file_in, 'r') as to_read:
        csv_r = csv.reader(to_read, delimiter = ';')
        
        first_line = csv_r.next()
        keys_typo = [int(x.split('_')[-1]) for x in first_line[1:len(first_line)]]
        
        for line in csv_r:
            ego = line[0]
            if not ego in ego_size:
                continue
            if not ego_size[ego] <= size_max:
                continue
            if ego in egos:
                ego = ego + '_%s' % (nb_par_ego.get(ego, 0) + 1)
                nb_par_ego[ego] = 1 + nb_par_ego.get(ego, 0)
            egos.append(ego)
            if not representativity:
                typo_vector[ego] = {int(keys_typo[i]) : float(line[i+1]) for i in range(len(keys_typo))}
            else:
                typo_vector[ego] = \
                {
                    int(keys_typo[i]) : float(line[i+1]) if float(line[i+1]) <= 1 \
                                        else 2-1/float(line[i+1])

                    for i in range(len(keys_typo))
                }
                    
                
    return egos, typo_vector, keys_typo

def get_overall_representativity():
    s = {x : 0 for x in DATA_MOTIFS[DATA_MOTIFS.keys()[0]]}
       
    for ego in DATA_MOTIFS:     
        for graphlet in DATA_MOTIFS[ego]:
            s[graphlet] += int(DATA_MOTIFS[ego][graphlet])
    
    all_graphlets = sum(s.values())
    return {x : s[x]/float(all_graphlets) for x in s}


def get_classe_representativity(list_of_egos):
    s = {x : 0 for x in DATA_MOTIFS[DATA_MOTIFS.keys()[0]]}
    for ego in list_of_egos:
        for graphlet in DATA_MOTIFS[ego]:
            s[graphlet] += int(DATA_MOTIFS[ego][graphlet])
       
    all_graphlets = float(sum(s.values()))
    local_representativities = {x : s[x]/all_graphlets for x in s}
    
    return {x : local_representativities[x] / float(OVERALL_REPR[x]) \
            if local_representativities[x] / float(OVERALL_REPR[x]) < 1
            else
                2 - 1 / (local_representativities[x] / float(OVERALL_REPR[x]))
            for x in local_representativities}

def create_dir(k_value):
    directory = '%s/Typo_%s' % (MAIN_DIR, k_value)
    if not 'Typo_%s' % k_value in os.listdir(MAIN_DIR):
        os.mkdir(directory)
    return directory

def kmeans_all_restreints(data):
    
    typo_par_ego_par_k = {ego : {} for ego in EGOS}

    for k_value in K_VALUES :
        s_max = 0
        labels_max = False
        for i in range(100):
            kmeans = KMeans(n_clusters = k_value)
            kmeans.fit(data)
            s = np.average(silhouette_samples(data, kmeans.labels_))
            if s > s_max:
                labels_max = kmeans.labels_
                s_max = s
            
        directory = create_dir(k_value)
        egos_par_classe = {}
        
        members_per_cluster = [[labels_max == k] for k in range(k_value)]
        
        for axe_2 in range(1, len(data.T)):
            for k in range(k_value):
                my_members = members_per_cluster[k][0]
                egos_par_classe[k] = []
                with open('%s/egos_classe_%s.csv' % (directory, k), 'w') as to_write:
                    csv_w = csv.writer(to_write, delimiter = ';')
                    csv_w.writerow(['ego_id'] + KEYS_TYPO)
                    for i in range(len(EGOS)):
                        if my_members[i]:
                            egos_par_classe[k].append(EGOS[i])
                            csv_w.writerow([EGOS[i]] + [TYPO_VECTOR[EGOS[i]][j] for j in KEYS_TYPO])
                            typo_par_ego_par_k[EGOS[i]][k_value] = k

        repr_par_classe, repr_par_classe_clustered = {}, {}
        for classe in egos_par_classe:
            repr_par_classe[classe] = get_classe_representativity(egos_par_classe[classe])
            
        with open('%s/kmeans_stats.csv' % directory, 'w') as to_write:
            csv_w = csv.writer(to_write, delimiter = ';')
            csv_w.writerow(['classe', 'nb', 'prop'] + [""] + KEYS_TYPO)
            classe_par_taille = [classe for classe in egos_par_classe]
            classe_par_taille.sort(key = lambda classe : len(egos_par_classe[classe]))
            for classe in classe_par_taille:
                nb = len(egos_par_classe[classe])
                prop = round(nb/float(len(EGOS)), 2)
                
                row = [classe, nb, prop] + [""]
                for key in KEYS_TYPO:
                    row.append(repr_par_classe[classe][key])                    
                csv_w.writerow(row)      

        print 'silhouette k = %s : %s' % (k_value, round(s_max,3))                 

    with open('%s/../typo_par_ego.csv' % directory, 'w') as to_write:
        csv_w = csv.writer(to_write, delimiter =';')
        csv_w.writerow(['ego'] + ['k = %s' % k for k in K_VALUES])
        for ego in typo_par_ego_par_k:
            csv_w.writerow([ego]+[typo_par_ego_par_k[ego][k_v] for k_v in K_VALUES])
            
corpus = 'facebook'
size_by_ego = {}
with open('../Data/size.csv', 'r') as to_read:
    csv_r = csv.reader(to_read, delimiter = ';')
    csv_r.next()
    for line in csv_r:
        size_by_ego[line[0]] = int(line[1])
        
for k in [5]:
    print k
    FILE_NAME = '%s_k%s' % (corpus, k)
    
    for size_max in [50,70,90,110,130,150]:
        DATA_MOTIFS = get_data(size_by_ego, size_max, 
                               file_name = '../Data/motifs_%s.csv' % FILE_NAME, 
                               representativity = False)[1]
        OVERALL_REPR = get_overall_representativity()

        file_name_split = FILE_NAME.split('_')
        motifs_same = 'motifs'
        for i in range(len(file_name_split)-1):
            motifs_same += '_%s' % file_name_split[i+1]

        MAIN_DIR = '../Results/%s_smax%s' % (FILE_NAME, size_max)
        print MAIN_DIR
        if not FILE_NAME + '_smax%s' % size_max in os.listdir('../Results/'):
            os.mkdir(MAIN_DIR)

        EGOS, TYPO_VECTOR, KEYS_TYPO = get_data(size_by_ego, size_max)
        DATA_CALC = pd.DataFrame.from_dict(TYPO_VECTOR, orient = 'index')

        K_VALUES = range(2, 8)
        kmeans_all_restreints(DATA_CALC)
```
