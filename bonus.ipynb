```python
folder = 'representativity_k4'
typo_1, typo_2 = 3, 5

import csv

table_passages = []

for i in range(typo_1):
    table_passages.append([])
    with open('../Results//{}/Typo_{}/egos_classe_{}.csv'.format(folder, typo_1, i), 'r') as to_read_1:
        csv_r_1 = csv.reader(to_read_1, delimiter = ';')
        csv_r_1.next()
        liste_ego_typo_1 = [line[0] for line in csv_r_1]
        
    for j in range(typo_2):
        nb_passages = 0
        with open('../Results//{}/Typo_{}/egos_classe_{}.csv'.format(folder, typo_2, j), 'r') as to_read_2:
            csv_r_2 = csv.reader(to_read_2, delimiter = ';')
            csv_r_2.next()
            for line in csv_r_2:
                if line[0] in liste_ego_typo_1:
                    nb_passages += 1
        table_passages[-1].append(nb_passages)     
        
table_passages
```

```python
import os
import csv
from igraph import *

def import_graph(ego):
    for folder in ['','all', 'csa', 'p5']:
        if os.path.isfile(os.path.expanduser('~/Graphs/%s/%s.gml' % (folder, ego))):
            return Graph.Read_GML(os.path.expanduser('~/Graphs/%s/%s.gml' % (folder, ego)))

def draw_graph(ego):
    graph = import_graph(ego)
    print ego
    
    all_to_rm = []
    for cc in graph.components():
        if len(cc) < 5:
            all_to_rm.extend(cc)
    graph.delete_vertices(all_to_rm)
            
    place = os.path.expanduser('~/Graphs/visu/%s.svg' % (ego))
    layout = graph.layout_fruchterman_reingold(repulserad = len(graph.vs)**3)
    graph.write_svg(place, layout = layout)
```

```python
with open('../Data/facebook_k5.csv', 'r') as to_read:
    csv_r = csv.reader(to_read, delimiter = ';')
    csv_r.next()
    for line in csv_r:
        draw_graph(line[0])
```
